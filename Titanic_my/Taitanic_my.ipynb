##  loading data

append the training set to the test set together.

```python
def combine_data():
    train = pd.read_csv("./data/train.csv")
    test = pd.read_csv("./data/test.csv")

    #extract and remove targets from training data
    targets = train.Survived
    train.drop("Survived",axis=1,inplace=True)

    #merge training data and testing data 
    #drop PassengerID since this is not a informative feature
    combined = train.append(test)
    combined.reset_index(inplace=True)    #attention: the index need to be reset
    combined.drop(["PassengerId","index"],axis=1,inplace=True)
    
    return combined
```

## processing the Age

in the first part ,we have seen that  the Age was missing 171 values.This is a large number(about 13% of the dataset).Simply replacing them with the mean or the median age might not be the best solution since the age might differ by groups and categories of passengers.

TO understand why,let's group our dataset by Sex,Title and passenger class and for each subset compute the median age.



```python
171/1300
```

 number of missing ages in train set

```python
#number of missing ages in train set
#combined[:891].Age.isnull().value_counts()
combined.iloc[:890].Age.isnull().sum()
#combined.iloc[890]
```

number of missing ages in train set

```python
combined[891:].Age.isnull().sum()
```

compute the meadians of groups by Sex,Pclass and Title

```python
grouped_train = combined[:891].groupby(["Sex","Pclass","Title"])
grouped_train.median()
```

```python
grouped_train.median().reset_index()
```

```python
grouped_train_median = grouped_train.median().reset_index()[['Sex', 'Pclass', 'Title', 'Age']]
```

This dataFrame will help us compute missing age values based on different categories.The meadian age can be differant based on the Sex,Pclass and Title put together.

Let's creat a fuction that fills in the missing age in combined based on theses different attributes.

```python
def fill_age(row):   #这里要以row为参数，对于null，需要整行信息
    #if row.Age == null ：  #错误，不能用null,是nan
    if np.isnan(row["Age"]):  #ok
        condition = ((grouped_train_median['Sex'] == row.Sex)&
        (grouped_train_median["Pclass"] ==row.Pclass) &
        (grouped_train_median['Title'] == row.Title))
        row.Age = grouped_train_median[condition]["Age"].values[0]
        return row['Age']
    else:
        return row["Age"]
    
        
    

combined["Age"] = combined.apply(fill_age,axis =1)
#combined['Age'] = combined.apply(lambda row1: fill_age(row1) if np.isnan(row1['Age']) else row1['Age'], axis=1)
```
